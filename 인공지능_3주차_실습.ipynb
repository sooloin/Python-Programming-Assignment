```python
#결측값 : 원치않는 값, 오류가 들어있는 값.
# 슛을 시도했을 때 몇 번 성공할 것인가?
# 따라서 의미있는 변수(컬럼)인 FGA, FGM을 뽑아내야 함.
```

#6. feature selection --> nbastat에서 FGA, FGM만 추출

```python
# FGA : 몇 개의 슛을 던지는지
# FGM : 몇 개의 슛을 성공하는지
# nbastat의 많은 column들 중에서 이 값들만 추출해서 사용
X = nbastat[['FGA']]
Y = nbastat[['FGM']]
print(X)
print(Y)
```

#7. 결측값을 처리

```python
# pandas에서 결측값을 해소하는 함수 : fillna
X = X.fillna(0) # 결측값은 0으로 채워주세요
Y = Y.fillna(0)
type(X)
```

7번을 통해 알 수 있는 것.

X와 Y는 pandas에서 제공하는 dataframe이지만 이 값들을 numpy에서는 사용할 수 없음.

그렇기에 실제로 X와 Y를 numpy에서 쓸 수 있는 배열로 바꿔줘야함 -> ndarray 함수 사용

#8. pandas의 dataframe --> numpy의 array로 변환

```python
X = (np.array(X)).reshape(m,1)
# X를 넘파이의 어레이로 바꿔주세요. 그리고 그 때의 형태를 mx1 형태로 바꿔주세요
Y = (np.array(Y)).reshape(m,1)
# Y를 넘파이의 어레이로 바꿔주세요. 그리고 그 때의 형태를 mx1 형태로 바꿔주세요

print(X)
type(X) # numpy에서 제공하는 자료구조로 이루어져있다는 것을 알 수 있음.
print(X.shape) # numpy에서 제공하는 배열이며 249 x 1 형태이다.
```

#9. 그려보기

```python
import matplotlib.pyplot as plt

plt.plot(X, Y, '.b')
# 반드시 .b여야 함. 그냥 b면 점 형태가 아니라 선 형태로 이어짐
plt.xlabel("FGA")
plt.ylabel("FGM")

# 선형적으로 증가하는 데이터임을 밑의 그래프를 통해 알 수 있음.
# 증가하는 직선의 식을 잘 세우면 얘네들을 잘 설명하는 인공지능 모델을 만들 수 있으리라 사료됨.
```

#10. 준비

```python
# 학습률(learning rate)
learning_rate = 0.0001

# 훈련 횟수 : epochs
n_iter = 200
```

#11. 초기화

```python
# theta와 theta의 미분(gradient)를 초기화 --> 0으로 초기화
theta = np.zeros((2, 1))
gradients = np.zeros((2,1))

print(theta.shape)
```

#12. 변수 설정

```python
#Xb를 설정 --> Xb = (1 X)의 결합
X0 = np.ones((m,1))
Xb = np.c_[X0, X]

print(Xb.shape)
print(theta.shape)
print(Y.shape)
# 입력 데이터에 바이어스를 추가하여 모델에 편향을 고려하는 전처리 과정을 수행합니다.
```

#13. 경사 하강법 수행(훈련)

```python
for i in range(n_iter): # n_iter값 높이면 데이터 묘사 정확도가 더 높아짐! 훈련횟수를 높여라!
  #1. Xb*theta --> Xb.dot(theta) => (249, 2) x (2, 1) --> (249, 1)
  #2. Xb*theta - Y --> Xb.dot(theta) - Y => (249, 1) - (249, 1) --> (249, 1)
  #3. (Xb*theta - Y)*Xb --> (249, 1) * (249, 2)
  #                     => Xb^T (Xb*theta - Y) --> (2, 249) * (249, 1) --> (2, 1)
  gradients = (1.0/m)*Xb.T.dot(Xb.dot(theta)-Y) # (2, 1)
  theta = theta - learning_rate * gradients
```

#14. 결과 가시화

```python
# (X, Y)의 데이터와 Y = theta_0 + X*theta_1
Y_pred = Xb.dot(theta)

plt.plot(X, Y_pred, color = 'red')
plt.plot(X, Y, '.b') # .b는 파란색 점을 표현하라는 의미, .r찍으면 빨간 점 나옴
```
